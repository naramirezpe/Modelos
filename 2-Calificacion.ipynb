# 2. Calificación : Predicción de etiquetas
### Etiquetas a predecir: 01 - 00 - 11 - 10

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import joblib
```

```python
df_02 = pd.read_csv('EntrenamientoModelo.csv', converters={'D1615': '{:0>2}'.format,'D1413': '{:0>2}'.format,'D1211': '{:0>2}'.format,
                      'D109': '{:0>2}'.format, 'D87': '{:0>2}'.format,'D65': '{:0>2}'.format,'D43': '{:0>2}'.format,
                      'D21': '{:0>2}'.format}, sep =';')
```

Renombrar las variables

```python
# Renombar variables, se entrena tomando como variables dependientes sd_2 y sd_1, 
# por lo tanto para calificacion sd_1 pasa a ser sd_3

df_03 = df_02.rename(columns={
                      'sd_16': 'sd_18',
                      'sd_15': 'sd_17',   
                      'sd_14': 'sd_16',
                      'sd_13': 'sd_15',
                      'sd_12': 'sd_14',
                      'sd_11': 'sd_13',
                      'sd_10': 'sd_12',
                      'sd_9': 'sd_11',
                      'sd_8': 'sd_10',   
                      'sd_7': 'sd_9',
                      'sd_6': 'sd_8',
                      'sd_5': 'sd_7',
                      'sd_4': 'sd_6',
                      'sd_3': 'sd_5',
                      'sd_2': 'sd_4',
                      'sd_1': 'sd_3'
                     })
```

```python
df_03
```

```python
modelo = joblib.load( 'modelo.joblib')
```

```python
X =  df_02[['sd_16', 'sd_15', 'sd_14', 'sd_13', 'sd_12', 'sd_11', 'sd_10', 'sd_9', 'sd_8', 'sd_7', 'sd_6', 'sd_5', 'sd_4', 'sd_3']]
```

```python
# Predicciones test
# ==============================================================================
Y_Resultado = modelo.predict(X)
```

```python
len(Y_Resultado)
```

```python
df_prediccion1 = pd.DataFrame(Y_Resultado, columns=['Prediccion01'])
```

```python
df_prediccion1
```

```python
df_prediccion1['Prediccion01'].unique()
```

```python
vector=list(df_prediccion1['Prediccion01'])
```

```python
col1=[]
col2=[]
for i in range(len(vector)):
    pareja=str(vector[i])
    nuevo=list(pareja)
    for j in range(len(nuevo)):
        if len(nuevo)==2:
            if j==0:
                col1.append(int(nuevo[j]))
            if j==1:
                col2.append(int(nuevo[j]))
        if len(nuevo)==1:
            col1.append(int(nuevo[j]))
            col2.append(int(nuevo[j]))
    
```

```python
df_prediccion1['sd_0']=col1
df_prediccion1['sd_+1']=col2
df_prediccion1
```
